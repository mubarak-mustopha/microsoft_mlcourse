```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

```python
X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
```

```python
class Regression:
  def __init__(self,x,y,model = linear_model.LinearRegression()):
    self.x = x
    self.y = y
    self.model = model

  def split_data(self,test_size=0.33):
    x = self.x
    y = self.y
    X_train, X_test, y_train, y_test = model_selection.train_test_split(x, y, test_size=0.33)
    return X_train, X_test, y_train, y_test

  def train_model(self,x_train,y_train):
    self.model.fit(x_train, y_train)

  def predict(self,X_test):
    pred = self.model.predict(X_test)
    return pred

  def plot_graph(self,X_test,y_test,y_pred,x_label,y_label):
    plt.scatter(X_test, y_test,  color='black')
    plt.plot(X_test, y_pred, color='blue', linewidth=3)
    plt.xlabel(str(x_label))
    plt.ylabel(str(y_label))
    plt.title(f'A Graph Plot Showing {y_label} against {x_label}')
    plt.show()
```

```python
features = ['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6']
for i , attribute in enumerate(features):
  x = X.copy()
  x = x[:, np.newaxis, int(i)]
  feature = Regression(x,y)
  X_train, X_test, y_train, y_test = feature.split_data()
  feature.train_model(X_train,y_train)
  y_pred = feature.predict(X_test)
  feature.plot_graph(X_test=X_test,y_test=y_test,y_pred=y_pred,x_label=attribute.upper(),y_label='Disease Progression')
```
