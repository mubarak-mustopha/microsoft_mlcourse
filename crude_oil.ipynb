```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
crude_df = pd.read_csv('data/crude_oil_Price.csv',index_col=0,parse_dates = True)
crude_df.head()
```

```python
crude_df.tail(3)
```

```python
crude_df.rename(columns={'Brent crude oil spot price, Monthly (dollars per barrel)':'Price'},inplace = True)
```

```python
import seaborn as sns
# Use seaborn style defaults and set the default figure size
```

```python
sns.set(rc={'figure.figsize':(11, 5)})
```

```python
crude_df['Price'].plot()
```

From the plot above we could see that the wasn'st really a significant change in oil price until about 2007/8 which then started to fluctuate after then.

```python
#we could zoom in the plot to between 2007 and 2010 to see better when the significant increase in oil price ocuured
crude_df.loc['2007':'2010','Price'].plot(marker='.', alpha=0.5)
```

```python
#there seemed to be a significant drop in price around 2020 ,let's check
crude_df.loc['2018':,'Price'].plot(marker='.', alpha=0.5)
```

Doesn't seem like there's any repeating pattern from the chart we just plotted but let's explore the data some more

```python
crude_df['Year'] = crude_df.index.year
crude_df['Month'] = crude_df.index.month_name()
crude_df['Week_day'] = crude_df.index.day_name()
```

```python
crude_df['Week_day'].value_counts()
```

```python
sns.boxplot(data=crude_df,x = 'Month', y='Price')
```

There's no clear variation in price over the months

```python
sns.boxplot(data=crude_df,x = 'Week_day', y='Price')
```
